HANDALING MISSING VALUE

# CUSTOMER_ID IS OUR PRIMARY_KEY SO WE HAE REMOVE THE ENTIRE ROW OF THAT MISSIN VALUE COLUM

```python
df.dropna(subset=['Customer_ID'], inplace = True)
df
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].astype(str).str.extract(r'(\d+\.?\d*)')
df['Age'] = df['Age'].astype(float)
```

```python
df.info()
```

in python we can store null value(na) in float but not in ineger

```python
age_median = int(df['Age'].median())
```

```python
age_median
```

```python
df['Age'].fillna(age_median, inplace= True)
```

```python
df['Age'].isnull().sum()
```

```python
df['Purchase_Amount'].fillna(df['Purchase_Amount'].median(),inplace=True)
```

```python
df['Purchase_Amount'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Feedback_Score'].fillna(df['Feedback_Score'].mode()[0], inplace= True)
```

```python
df.isnull().sum()
```

```python
for col in ['Gender','City','Country']:
    df[col].fillna(df[col].mode()[0], inplace= True)
```

```python
df['Last_Purchase_Date'].ffill(inplace= True)
```

```python
df.isnull().sum()
```

# handaling outliers

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Box Plot for Age
plt.subplot(1,3,1)
plt.boxplot(df['Age'].dropna())
plt.title('Age')
plt.ylabel('Values')

# Box Plot for Purchase Amount
plt.subplot(1,3,2)
plt.boxplot(df['Purchase_Amount'].dropna())
plt.title('Purchase Amount')

# Box Plot for Feedback Score
plt.subplot(1,3,3)
plt.boxplot(df['Feedback_Score'].dropna())
plt.title('Feedback Score')

plt.tight_layout()
plt.show()
```

there exist outliers in age and purchas amount

```python
from scipy import stats
import numpy as np
```

```python
z_scores= np.abs(stats.zscore(df[['Age','Purchase_Amount']]))
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df[['Age', 'Purchase_Amount']]))

outliers = df[(z_scores > 3).any(axis=1)]

outliers
```

```python
df_clean =df[~(z_scores >= 3).any(axis=1)]
```
